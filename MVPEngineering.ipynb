# MVP Engineering

## Initial Results 

Below is a plot of Steph Curry's year to year statistics for Offensive Win Shares, Defensive Win Shares, and Total Win Shares. 

This is an example of what will be imbedded in a Streamlit application. The user will be able to search a player and a specific stat or stats and then view a plot of this combination. 



*An interesting note is that in the year 2015-2016, Steph Curry had career highs in OWS, DWS, and Total Win Shares. This is the year he was voted as the unanimous MVP of the NBA. Unfortunately the Warriors lost in the finals to the Cavs.*
